```python
# PREPROCESSING DATA
import statistics as st
from math import isnan
from itertools import filterfalse   # fixed here

data = [85.3, float('NaN'), 92.1, 78.6, float('NaN'), 89.4, 76.8, 94.2]

# Number of missing values
print("Missing values:", sum(map(isnan, data)))

# strip NaN values using 'filterfalse'
clean = list(filterfalse(isnan, data))
print("Clean data:", clean)

# sorting the cleaned data
print("Sorted clean data:", sorted(clean))

# Statistics
print('median =', st.median(clean))
print('median low =', st.median_low(clean))
print('median high =', st.median_high(clean))
print('mode (numbers) =', st.mode([1, 1, 2, 3, 3, 3, 4]))
print('mode (colors) =', st.mode(["red", "blue", "blue", "red", "green", "red", "red"]))
print('multimode =', st.multimode('aabbbbbccdddddeeeffffgg'))

# variance
data = [22, 7, 10, 13, 14, 14, 32, 19, 22, 24]
print("variance =", st.variance(data))
```

```python
#scipy implementation
import numpy as np
from scipy import special
from scipy import linalg
from scipy import integrate
from scipy.misc import derivative
from math import sqrt
print("## SciPy Special Functions ##")

# Calculate 10**2
a = special.exp10(2)
print(f"special.exp10(2) = {a}")

# Calculate 2**3
b = special.exp2(3)
print(f"special.exp2(3) = {b}")

# Sine of 90 degrees
c = special.sindg(90)
print(f"special.sindg(90) = {c}\n")
```

```python
print("## Linear Algebra ##")
# Inverse of a 2x2 matrix
print("# Inverse of a 2x2 matrix")
a_inv = np.array([[1, 33], [3, 4]])
b_inv = linalg.inv(a_inv)
print(b_inv)
print("-" * 20)

# Determinant of a 3x3 matrix
print("# Determinant of a 3x3 matrix")
mat = np.array([[1, 33, 2], [3, 2, 1], [1, 2, 1]])
mat_det = linalg.det(mat)
print(mat_det)
print("-" * 20)

# Inverse of a 3x3 matrix
print("# Inverse of a 3x3 matrix")
# Using the same matrix 'mat' from the determinant example
mat_inv = linalg.inv(mat)
print(mat_inv)
print("-" * 20)
```

```python
# Eigenvalues and eigenvectors
print("# Eigenvalues and Eigenvectors")
arr = np.array([[33, 34], [36, 53]])
eg_val, eg_vect = linalg.eig(arr)
print("Eigenvalues:")
print(eg_val)
print("Eigenvectors:")
print(eg_vect)
print("\n")
```

```python

print("## Calculus ##")

# Derivative of a function
print("# Derivative of a function")
def my_function(x):
    return x**2 + x + 1
deriv = derivative(func=my_function, x0=5)
print(f"The derivative of x**2 + x + 1 at x=5 is: {deriv}")
print("-" * 20)

# Single integration
print("# Single Integration")
f_single = lambda x: x**2
integration_single = integrate.quad(f_single, 0, 1)
print(f"Result of single integration of x**2 from 0 to 1:")
print(integration_single)
print("-" * 20)

# Double integration
print("# Double Integration")
f_double = lambda x, y: 64 * x * y
# Lower limit of the inner integral (p)
p = lambda x: 0

q = lambda y: sqrt(1 - 2 * y**2)
integration_double = integrate.dblquad(f_double, 0, 2/4, p, q)
print(f"Result of double integration:")
print(integration_double)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load dataset
dataset = pd.read_csv("C:/Users/jaide/Downloads/kaggle_house/Housing.csv")

# Features and target
X = dataset.drop("price", axis=1)   # all except price
y = dataset["price"].values         # target

# Convert categorical features into numbers
X = pd.get_dummies(X, drop_first=True)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)

# Train model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predictions
y_pred = regressor.predict(X_test)

# Plot Actual vs Predicted
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linewidth=2)  # Diagonal line
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted House Prices')
plt.grid(True)
plt.show()
```

```python
# 1. Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# 2. Load the dataset
df = pd.read_csv("C:/Users/jaide/Downloads/kaggle_house/Housing.csv")

# --- PREPROCESSING STEP ---
# Convert 'yes'/'no' columns to 1/0
categorical_cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
for col in categorical_cols:
    df[col] = df[col].map({'yes': 1, 'no': 0})

# Convert 'furnishingstatus' into numerical columns using one-hot encoding
dummies = pd.get_dummies(df['furnishingstatus'], drop_first=True, dtype=int)
df = pd.concat([df, dummies], axis=1)
df = df.drop('furnishingstatus', axis=1)


# 3. Define features (X) and target (y)
# Now we can use all the relevant columns
X = df[['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad', 'guestroom',
        'basement', 'hotwaterheating', 'airconditioning', 'parking', 'prefarea',
        'semi-furnished', 'unfurnished']]

y = df['price']
```

```python
# 4. Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)

# 5. Create and train the linear regression model
lm = LinearRegression()
lm.fit(X_train, y_train)

# 6. Make predictions
predictions = lm.predict(X_test)
```

```python
# --- NEW: Preview the train and test datasets ---
print("--- Data Split Preview ---")
print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print("\nFirst 5 rows of the training features (X_train):")
print(X_train.head())
print("\n--------------------------\n")
```

```python
# 7. Visualize results
plt.figure(figsize=(8, 6))
plt.scatter(y_test, predictions, color='orange', edgecolor='k', alpha=0.7)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs. Predicted Prices')
plt.grid(True)
plt.show()

# 8. Evaluate the model
print('Model Evaluation Metrics:')
mae = metrics.mean_absolute_error(y_test, predictions)
mse = metrics.mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)

print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Square Error (RMSE): {rmse}')
```

```python
X = iris.iloc[:,:-1]
y = iris.species
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33,random_state=42)
treemodel = DecisionTreeClassifier()
treemodel.fit(X_train,y_train)
plt.figure(figsize = (15,10))
tree.plot_tree(treemodel,filled = True)
```
